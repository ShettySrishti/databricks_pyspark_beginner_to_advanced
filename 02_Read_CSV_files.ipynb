### Read Single CSV File

```python
df = spark.read.options(header=True, inferSchema=True).csv("dbfs:/data/input/emp.csv")
display(df)
print(df.count())
```

### Read Multiple CSV Files

```python
df = spark.read.options(header=True, inferSchema=True).csv(["dbfs:/data/input/emp.csv", "dbfs:/data/input/emp_new.csv"])
display(df.limit(5))
print(df.count())
```

### Read all CSV files under a Folder

```python
# all CSV files in the directory should have the same schema and there should be no nested folders
df = spark.read.options(header=True, inferSchema=True).csv("dbfs:/FileStore/tables") 
display(df)
print(df.count())
```

### Create Own Schema

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType

schema_defined = StructType(
    [StructField('deptid', IntegerType(), True),
    StructField('deptcode',StringType(), True),
    StructField('deptname', StringType(), True)]
)
```

```python
df = spark.read.schema(schema_defined).options(header=True).csv("dbfs:/data/input/dept.csv")
display(df.limit(5))
```

```python
schema_alternate = 'dept_id INTEGER, dept_code STRING, dept_name STRING'
```

```python
df = spark.read.schema(schema_alternate).options(header=True).csv("dbfs:/data/input/dept.csv")
display(df.limit(5))
```
