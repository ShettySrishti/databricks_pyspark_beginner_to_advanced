### PySpark Approaches

Approach1 : TimeStamp + Table 

```python
df = spark.read.format("delta")\
    .option("timestampAsOf", "2025-11-29T15:48:40.000+00:00")\
    .table("scd2demo")

df.display()
```

Approach2 : TimeStamp + Path

```python
df = spark.read.format("delta")\
    .option("timestampAsOf", "2025-11-29T15:48:40.000+00:00")\
    .load("/FileStore/tables/scd2demo")

df.display()
```

Approach3 : Version + Path

```python
df = spark.read.format("delta")\
    .option("versionAsOf", 0)\
    .load("/FileStore/tables/scd2demo")

df.display()
```

Approach4 : Version + Table

```python
df = spark.read.format("delta")\
    .option("versionAsOf", 1)\
    .table("scd2demo")

df.display()
```
