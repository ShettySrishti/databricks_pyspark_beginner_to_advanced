- These DELETE And UPDATE operations are going to create very small partition files which in turn is not good for performance.
- In order to avoid creation of smaller files, we will use `OPTIMZE` command.
- `OPTIMIZE` will merge smaller active files into one single partition 1GB (max default size) file. 
- Check Catalog for the same

```python
%sql
OPTIMIZE scd2demo -- The future operations will use this combined active file which will increase the performance
```
