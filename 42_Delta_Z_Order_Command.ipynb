```python
%sql
OPTIMIZE scd2demo
ZORDER BY pk1
```

- Z-ordering is a data clustering technique in Delta Lake that organizes data files to colocate related column values, improving the performance of queries that filter on those columns.
- Eg: Z-order in scd2demo clusters data files by the pk1 column to speed up queries that filter on pk1
