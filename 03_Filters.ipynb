### Create DataFrame

```python
employee_data = [(10, "Srishti Shetty", "1998", "100", "F", 8000),
                 (20, "Aish Rai", "2002", "200", "M", 2000),
                 (30, "Rishabh Alva", "2010", "100",None, 6000),
                 (40, "Aarvi Bhandhary", "1996", "400", "F", 7000),
                 (50, "Pooja Hegde", "2008", "500", "F", 5000),
                 (60, "Sunil Shetty", "1997", "400", "M", 3000),
                 (70, "Sid Rai", "2010", "600", "M", 5000)
                 ]

employee_schema = ["employee_id", "name", "doj", "employee_dept_id", "gender","salary"]

employee_DF = spark.createDataFrame(data = employee_data, schema = employee_schema)
display(employee_DF)
```

```python
print(employee_DF.columns)
```

```python
from pyspark.sql.functions import col
display(employee_DF.filter(col("salary")<=5000)) # ==, >=, <=, >, <, !=

display(employee_DF.filter(employee_DF.salary<=5000))
```

```python
employee_DF.filter((employee_DF.salary<=5000) & (employee_DF.doj<= "2010")).show() # & , |
employee_DF.filter((employee_DF.salary<=5000) | (employee_DF.doj== "2010")).show() 
```

```python

employee_DF.filter(employee_DF.name.startswith("S")).show()  # startswith, endswith, contains
employee_DF.filter(employee_DF.name.contains("Aarvi")).show() 
employee_DF.filter(employee_DF.name.endswith("Rai")).show() 
```

```python
employee_DF.filter(employee_DF.gender.isNull()).show() 
employee_DF.filter(employee_DF.gender.isNotNull()).show() 
```

```python
employee_DF.filter(employee_DF.salary.isin("2000", "5000")).show() 
```

```python
employee_DF.filter(~ employee_DF.salary.isin("2000", "5000")).show() # ~ isin --> is not in
```

```python
employee_DF.filter(employee_DF.name.like("A%")).show() # %char%, char%, %char
```
