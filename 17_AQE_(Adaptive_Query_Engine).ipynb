### Adaptive Query Execution (AQE) features in Spark

**Dynamic Join Selection:**
- Spark can change the join strategy (e.g., from sort-merge join to broadcast join) at runtime based on actual data statistics.

**Dynamic Partition Coalescing:**
- Spark can merge small shuffle partitions into larger ones during execution to avoid small file problems and improve performance.

**Dynamic Skew Join Handling:**
- Spark can detect data skew at runtime and optimize join execution by splitting skewed partitions.

**Dynamic Optimizations:**
- AQE can optimize query plans based on runtime metrics, such as changing the number of reducers or re-optimizing physical plans.

**Reduced Need for Manual Tuning:**
- AQE reduces the need for users to manually tune shuffle partitions and join strategies, as Spark adapts automatically.

**Summary:**
- AQE makes Spark queries more efficient and robust by adapting execution plans dynamically based on actual data and runtime statistics.
