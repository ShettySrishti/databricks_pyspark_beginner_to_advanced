```python
dbutils.widgets.dropdown("DropDown", "1", [str(x) for x in range(1, 6)]) # value choices must be string
```

```python
dbutils.widgets.combobox("ComboBox", "1", [str(x) for x in range(1, 6)]) # similar to dropdown
```

```python
dbutils.widgets.multiselect("Product", "Camera", ("Camera", "AC", "SmartPhone"))
```

```python
dbutils.widgets.remove("ComboBox") # To remove one widget
```

```python
dbutils.widgets.removeAll() # To remove all widgets
```
