### Read CSV file without any option

```python
df_no_option = spark.read.csv("/data/input/employee.csv")
```

**Only One Spark job created!**

### Read CSV file with inferSchema option

```python
df_inferSchema = spark.read.option("inferSchema", True).option("header", True).csv("/data/input/employee.csv")
```

**Two Spark Jobs Created!!**

```python
df_inferSchema.show()
```

```python
df_inferSchema.printSchema()
```

### Read CSV file with Defined Schema

```python
from pyspark.sql.types import StringType, StructField, StructType, IntegerType, DateType
schema = StructType([StructField("employee_id", StringType(), True),\
            StructField("department_id", IntegerType(), True),\
            StructField("name", StringType(), True),\
            StructField("age", IntegerType(), True),\
            StructField("gender", StringType(), True),\
            StructField("salary", IntegerType(), True),\
            StructField("hire_date", DateType(), True)\
              ])
```

```python
df_defined_schema = spark.read.schema(schema).csv("/data/input/employee.csv")
```

**No Spark Jobs created!!!!**
