### Create Sample Dataframe with NULL values

```python
data = [
    ("John Smith", "Math", 85, "Pass", 90),
    ("Maria Garcia", "Science", None, "Fail", 80),
    ("James Miller", "English", 78, None, 85),
    ("Linda Johnson", "History", 92, "Pass", None),
    ("Robert Brown", None, 88, "Pass", 95),
    ("Emily Clark", "Math", 91, "Pass", 100),
    ("Michael Lee", "Science", 76, "Fail", None),
    ("Sophia Turner", "English", None, "Pass", 88),
    ("William Harris", "History", 83, None, 92),
    ("Olivia Walker", "Math", 89, "Pass", None),
    (None, None, None, None, None)  # Entirely None row
]

columns = ["name", "subject", "marks", "status", "attendance"]

df = spark.createDataFrame(data, columns)
df.show()
```

### Find NULL Occurences of Each Column in Dataframe

```python
from pyspark.sql.functions import count, col, when

result = df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns])
display(result)
```
