### CReate a Sample Dataframe

```python
data = [
    ("Alice", 85, 78, 92, 88, 90),
    ("Bob", 76, 81, 79, 85, 80),
    ("Charlie", 90, 88, 85, 87, 89),
    ("David", 65, 70, 72, 68, 75),
    ("Eva", 95, 92, 94, 96, 93)
]

schema = ["Student", "Sub_1", "Sub_2", "Sub_3", "Sub_4", "Sub_5"]

df =spark.createDataFrame(data, schema)
df.display()
```

### Greatest of Columns

```python
from pyspark.sql.functions import greatest

df_greatest = df.withColumn("Greatest", greatest("Sub_1", "Sub_2", "Sub_3", "Sub_4", "Sub_5"))
df_greatest.display()
```

### Least of Columns

```python
from pyspark.sql.functions import least

df_least = df.withColumn("Least", least("Sub_1", "Sub_2", "Sub_3", "Sub_4", "Sub_5"))
df_least.display()
```

### Max Of Rows

```python
from pyspark.sql import functions as F

df.agg(
    F.max("Sub_1").alias("Max_Sub_1"),
    F.max("Sub_2").alias("Max_Sub_2"),
    F.max("Sub_3").alias("Max_Sub_3"),
    F.max("Sub_4").alias("Max_Sub_4"),
    F.max("Sub_5").alias("Max_Sub_5")
).show()
```

```python
df.agg({'Sub_1': 'max'}).display()
```

```python
df.agg({'Sub_1': 'max', 'Sub_2':'max', 'Sub_3':'max'}).display()
```

### Min Of Rows

```python
df.agg({'Sub_1': 'min', 'Sub_2': 'min'}).toDF('Min_Sub_1', 'Min_Sub_2').display()
```
