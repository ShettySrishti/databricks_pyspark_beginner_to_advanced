### Create a sample Dataframe

```python
empDF = [
    ("John Smith", [85, 87, 88, 90, 93, 91]),
    ("Maria Garcia", [78, 79, 80, 88, 85, 87]),
    ("James Miller", [92, 90, 91, 95, 94, 96]),
    ("Linda Johnson", [80, 82, 83, 84, 86, 85]),
    ("Robert Brown", [88, 89, 87, 91, 92, 90])
]

array_schema = ["Name", "Array_col"]
arrayDF = spark.createDataFrame(empDF, array_schema)
display(arrayDF)
```

### Apply Array Sort

```python
from pyspark.sql import functions as F

sortedDF = arrayDF.withColumn("Sorted", F.array_sort("Array_col"))
sortedDF.show(truncate=False)
```
