### Create Employee Dataframe

```python
employee_data = [(10, "Srishti","Shetty", "1998", "100", "F", 8000),
                 (20, "Aish", "Rai", "2002", "200", "M", 2000),
                 (30, "Rishabh", "Alva", "2010", "100",None, 6000),
                 (40, "Aishanya",  "Bhandhary", "1996", "400", "F", 7000),
                 (50, "Pooja",  "Hegde", "2008", "500", "F", 5000),
                 (60, "Sunil", "Shetty", "1997", "400", "M", 3000),
                 (70, "Sumit", "Rai", "2010", "600", "M", 5000)
                 ]

employee_schema = ["employee_id", "first_name", "last_name", "doj", "dept_id", "gender","salary"]

emp_DF = spark.createDataFrame(data = employee_data, schema = employee_schema)
display(emp_DF)
```

### Create Department Dataframe

```python
department_data = [("HR", 100),
                   ("Supply", 200),
                   ("Sales", 300),
                   ("Stock", 400)]

department_schema = ["dept_name", "dept_id"]
dept_DF= spark.createDataFrame(department_data, department_schema)
display(dept_DF)
```

### Inner Join

```python
# df_join = emp_DF.join(dept_DF, "dept_id", "inner")  # alternate if key is same
df_join = emp_DF.join(dept_DF, emp_DF.dept_id==dept_DF.dept_id, "inner")
df_join.show()
```

### Full Outer Join

```python

df_join = emp_DF.join(dept_DF, "dept_id", "full") # outer, full_outer, fullouter
df_join.show()
```

### Left Outer Join/Left Join

```python
df_join = emp_DF.join(dept_DF, "dept_id", "left") # left_outer, leftouter
df_join.show()
```

### Right Outer Join / Right Join

```python
df_join = emp_DF.join(dept_DF, "dept_id", "right") # right_outer, rightouter
df_join.show()
```

### Left Semi Join

```python
df_join = emp_DF.join(dept_DF, "dept_id", "semi") # left_semi, leftsemi
df_join.show()
```

### Left Anti Join

```python
df_join = emp_DF.join(dept_DF, "dept_id", "anti") # left_anti, leftanti
df_join.show()
```
