### Create dataframe with array column

```python
array_appliance = [('Srishti', ['Smart Phone', 'Oven', 'AC', 'Smart TV']),
                   ('Aarvi', None),
                   ('Aish', ['Refridgerator', 'Washing Machine', 'AC']),
                   ('Jenny', ['Grinder', 'AC', 'Smart TV']),
                   ('Rohan', ['Oven', 'AC', None])]

df_app = spark.createDataFrame(array_appliance, ['Name', 'Appliances'])
display(df_app)
```

### Create dataframe with map column

```python
map_appliance = [
    ('Srishti', {'Smart Phone': 'Pixel', 'Oven': 'LG', 'AC': 'Philips', 'Smart TV': 'LG'}),
    ('Aish', {'Refridgerator': 'Whirlpool', 'Washing Machine': 'Bosch', 'AC': 'Daikin'}),
    ('Jenny', {'Grinder': 'Preethi', 'AC': 'Daikin', 'Smart TV': 'Samsung'}),
    ('Rohan', {'Oven': 'LG', 'AC': 'Voltas'}),
    ('Aarvi', None)
]

df_brand = spark.createDataFrame(map_appliance, ['Name', 'Brand'])
display(df_brand)
```

```python
from pyspark.sql.functions import explode, explode_outer, posexplode, posexplode_outer
```

### Explode Array Field

**_## Row with None is ignored by explode._**

```python
df2= df_app.select(df_app.Name, explode(df_app.Appliances)) # Will not display person with no appliances --> null value
display(df2)
```

### Explode Map Field

```python
df3= df_brand.select(df_brand.Name, explode(df_brand.Brand)) # Will not display person with no appliances --> null value
display(df3)
```

### Explode Outer to consider None rows 

```python
display(df_app.select(df_app.Name, explode_outer(df_app.Appliances)))
```

### Positional Explode

```python
display(df_app.select(df_app.Name, posexplode(df_app.Appliances))) 
```

### 'posexplodeouter' will consider Rows with None and assign 'pos'= null to that row
