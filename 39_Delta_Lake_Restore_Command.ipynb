### Create a Delta Table Instance

```python
from delta import *
tgtTable = DeltaTable.forPath(spark, "/FileStore/tables/scd2demo")
```

### List down Version History

```python
tgtTable.history().display()
```

Perform a DELETE operation so that we can restore it to its default version later

```python
%sql
DELETE FROM scd2demo WHERE active_status = 'N'
```

```python
tgtTable.history().display()  # Ignore 3 and 4 versions
```

```python
%sql
SELECT * FROM scd2demo
```

### RESTORE using Version Number

```python
tgtTable.restoreToVersion(2) # Restore table to its oldest version
```

```python
%sql
SELECT * FROM scd2demo -- Restored
```

```python
tgtTable.history().display()  
```

### RESTORE using TimeStamp

```python
tgtTable.restoreToTimestamp("2025-11-29T15:48:40.000+00:00") # version 1 three rows
```
