### Snappy vs GZip

```python
csvDF = spark.read.option("header", True).option("inferSchema", True).csv("dbfs:/FileStore/dbutils_test/superstore.csv")
csvDF.show(n=5)
```

```python
csvDF.write.option("compression", "snappy").parquet("dbfs:/FileStore/dbutils_test/write_file/snappy_parquets")
```

```python
%fs
ls dbfs:/FileStore/dbutils_test/write_file/snappy_parquets
```

```python
csvDF.write.option("compression", "gzip").parquet("dbfs:/FileStore/dbutils_test/write_file/gzip_parquets")
```

```python
%fs
ls FileStore/dbutils_test/write_file/gzip_parquets
```
