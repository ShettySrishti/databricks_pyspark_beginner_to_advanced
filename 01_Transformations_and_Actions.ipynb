### Read source data to create dataframe

```python
df = spark.read.options(header=True, inferSchema=True, sep=",").csv("dbfs:/data/input/dept.csv")
display(df)
```

### Perform Narrow Transformation1: Filter

```python
df1 = df.filter(df.department_name=='Data and Analytics')
df1.show()
```

### Perform Narrow Transformation2: Filter

```python
df2 = df.filter(df.department_code.isin('DNA','Finance'))
df2.show()
```

### Perform Narrow Transformation3: Union

```python
df3 =df1.union(df2)
df3.show()
```

### Perform Wide Transformation: GroupBy

```python
df4 = df.groupBy("department_name").count()
```

### Calling Action Collect

```python
df4.collect()
```
