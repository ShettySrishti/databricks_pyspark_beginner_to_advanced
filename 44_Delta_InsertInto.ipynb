```python
data = [("ABC pvt. ltd.", "Q1", 2000),
        ("XYZ pvt. ltd.", "Q1", 5000),
        ("KLM pvt. ltd.", "Q1", 3000)]
col = ["Company", "Quarter", "Revenue"]

df = spark.createDataFrame(data, col)
df.display()
```

```python
df.write.saveAsTable("fact_revenue")
```

### Retrieve all the records from the table

```python
spark.table("fact_revenue").display() # Approach 1
```

```python
spark.sql("SELECT * FROM fact_revenue").display() # Approach 2
```

```python
%sql
SELECT * FROM fact_revenue -- Approach 3
```
