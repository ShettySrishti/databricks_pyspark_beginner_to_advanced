### Read Sample dataframe

```python
df = spark.read.option("inferSchema", True).option("header", True).csv("/data/input/dept.csv")
display(df)
```

### Number of records in dataframe

```python
df.count()
```

### Default Partition Count

```python
print(df.rdd.getNumPartitions())
```

### Number of Records Per Partition

```python
from pyspark.sql.functions import spark_partition_id

df.withColumn("partitionid", spark_partition_id()).groupBy("partitionid").count().show()
```

### Repartition the Dataframe to 1

```python
df1 = df.repartition(1)
```

```python
print(df1.rdd.getNumPartitions())
```

```python
df1.withColumn("partitionid", spark_partition_id()).groupBy("partitionid").count().show()
```
