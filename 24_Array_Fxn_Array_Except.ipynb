### Create a Sample Dataframe

```python
empDF = [
    ("John Smith", [85, 87, 88], [90, 88, 85]),
    ("Maria Garcia", [78, 79, 80], [88, 78, 87]),
    ("James Miller", [92, 90, 91], [91, 90, 92]),
    ("Linda Johnson", [80, 82, 83], [84, 86, 85]),
    ("Robert Brown", [91, 90, 87], [91, 92, 90])
]

array_schema = ["Name", "Array_1", "Array_2"]
arrayDF = spark.createDataFrame(empDF, array_schema)
display(arrayDF)
```

### Apply Array Except

```python
from pyspark.sql import functions as F

exceptDF = arrayDF.withColumn("Exception", F.array_except("Array_1", "Array_2"))
display(exceptDF)
```
